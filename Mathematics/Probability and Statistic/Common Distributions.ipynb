# Common Distributions
## Geometric Distribution
$$
X \sim \text{Geom}(p)
$$
### Probability Mass Function
$$
P(X=k)=(1-p)^{k-1}p
$$
$X$ is the number of trials needed to see the first success in a sequence of independent trials with success probability $p$ .

#### $p=0.2$

```python
P = (1-p)**(k-1)*p
#plot(P.subs(p,0.5), (k, 0, 10))
discrete_k = range(1, 11, 1)
discrete_P = [P.subs(p, 0.2).subs(k, ki) for ki in discrete_k]
plt.bar(discrete_k, discrete_P)
plt.show()
P.subs(p,0.2)
```

## Exponential Distribution
$$
X \sim \text{Exp}(\lambda)
$$
Probability Density Function:
$$
f(x) = 
\begin{cases}
\lambda e^{-\lambda x} &, x \geq 0\\
0&, x<0
\end{cases}
$$
Cumulative Distribution Function:
$$
F(t)=\int_{-\infty}^tf(x)dx=1-e^{-\lambda t}
$$
$$
P(X\leq t)=F(t)=1-e^{-\lambda t}
$$
$$
P(X>t)=1-F(t)=e^{-\lambda t}
$$
$\lambda$ is the intensity or rate of Poisson Process (the incident happens $\lambda$ times per unit time in average). $X$ is the first time that the incident happens.

### Probability Density Function

```python
f = λ*exp(-λ*x)
plot(f.subs(λ, 1), (x, 0, 10))
f.subs(λ, 1)
```

### Cumulative Distribution Function

```python
F = 1-exp(-λ*t)
plot(F.subs(λ, 1), (t, 0, 10))
F.subs(λ, 1)
```

## Poison Distribution
$$
X \sim \text{Poison}(\mu)
$$
### Probability Mass Function
$$
P(X=k)=e^{-\mu}\frac{\mu^{k}}{k!}
$$
$X$ is the number that the trials succeed while $\mu$ is the average success number (expectation).

The connections between:

* Poisson Distribution and Exponential Distribution 
* $\mu$ and $\lambda$
  
is such that:
$$
\mu = \lambda \cdot t
$$
,which is the important relationship in Poisson Process.

#### $\mu = 0.2$

```python
P = exp(-μ)*μ**(k)/factorial(k)
discrete_P = [P.subs(μ, 0.2).subs(k, ki) for ki in discrete_k]
plt.bar(discrete_k, discrete_P)
plt.show()
P.subs(μ, 0.2)
```

# Relationship Among Them
## In Poisson Process
* **In a given time interval**, the number of incident happens can be modeled with a Poisson random variable $X$.
  * $X \sim \text{Poison}(\mu)$
  * $P(X=k)=e^{-\mu}\frac{\mu^{k}}{k!}$
  * $X$ has the one dimension.
* Consider the time interval **between two sequential incidents**, the duration of waiting can be modeled with a Exponential random variable $X$.
  * $X \sim \text{Exp}(\lambda)$
  * $f(x) = 
\begin{cases}
\lambda e^{-\lambda x} &, x \geq 0\\
0&, x<0
\end{cases}$
  * $X$ has the dimension of time, and therefore can also be denoted as $T$
* The relationship between two parameters
  * $\mu = \lambda \cdot t$
  * $\mu$ has the dimension of one.
  * $\lambda$ has the dimension of rate (number per unit time).
* Exponential Distribution is the continuos version of Geometry Distribution.
  * That means take trials very frequently.
    * $n \to \infty$, $n$ is the trials number per unit time.
  * And probability succeed every trial very small given constant $\lambda$.
    * $p = \frac{\lambda}{n}$
